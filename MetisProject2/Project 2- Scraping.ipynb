```python
# Use requests to pull data
response = requests.get(url)
```

```python
# Use requests to pull data
response = requests.get(url)

# Check to make sure we are getting response
response.status_code # status code = 200 => OK

# Check out the text
response.text

page = response.text

from bs4 import BeautifulSoup

soup = BeautifulSoup(page, "lxml")

print(soup.prettify)
```

```python
# Get tables, find the one I want (-1)
my_table = [table for table in soup.find_all('table', class_='wikitable')][-1]
```

```python
# Find a (hyperlinks) because the names are hyperlinks
links = my_table.find_all('a')
print(links)
```

```python
# Look through links to extract names
attractions = []
for link in links:
    attractions.append(link.get('title'))
print(attractions)
```

```python
cleaned_attractions = []
for x in attractions:
    if type(x) == str:
        cleaned_attractions.append(x)
    
```

```python
cleaned_attractions
```

```python
# Send data to pd df
attractions_df = pd.DataFrame()
attractions_df['Attractions'] = cleaned_attractions
```
